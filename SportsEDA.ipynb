## Wins per Team

```python
win=data['winner'].value_counts().sort_index()
win
```

```python
teams=data['winner'].unique()
teams.sort()
```

```python
teams
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(20,9))
sns.barplot(x=teams,y=win)
plt.xticks(rotation=90)
plt.xlabel("Team Name")
plt.ylabel("Number of wins")
plt.title("Wins Per Team")
plt.show()
```

## Most Player Of Match

```python
mom=data['player_of_match'].value_counts().head(20)
mom
```

```python
pqr=mom.index
qrs=mom.values
```

```python
plt.figure(figsize=(20,9))
sns.barplot(x=pqr,y=qrs)
plt.xticks(rotation=65)

plt.show()
```

## Winning Trend based on Batting First or Batting Second

```python
data['wins']=np.where(data['win_by_runs']>0,'Team Batting First','Team Batting Second')
```

```python
data.head()
```

```python
data['wins'].value_counts().plot(kind = 'pie', figsize = (15,8), fontsize=10,autopct='%1.2f%%',textprops = {"fontsize":15})
```

## Toss to Win Comparision

```python
data['toss_comparision']=np.where(data['toss_winner']==data['winner'],'Match In favour of toss','Match not in favour of toss')
```

```python
data['toss_comparision'].value_counts().plot(kind = 'pie', figsize = (15,8), fontsize=10,autopct='%1.2f%%',textprops = {"fontsize":15})
```

## Mumbai Indians Winning Trend

```python
x=data[data['winner']=='Mumbai Indians']
```

```python
x
```

```python
plt.subplots(figsize=(13,7))
x.groupby(['season']).count()['id'].plot(color = 'red')
plt.xticks(rotation=0)
plt.title("Trend of wins for mumbai indians")
plt.xlabel("Years")
plt.ylabel("Total wins")
```
